# 実験状況

## 放射光源（SR）
ビームラインのアンジュレータから放射光が出力されます．
## モノクロメータ（結晶分光器）
通常のビームラインモノクロメータはいくつかの電子ボルトの帯域幅でSRのエネルギーを決定します．このエネルギーの幅はトランスミッターやスキャッターで分析する超微細構造よりもはるかに広いです．
## トランスミッター（測定試料）
サンプルまたはエネルギー基準物質が配置され放射光はこれを透過します．トランスミッターの後ろにはサンプルまたはエネルギー基準物質が配置されます．この二つのうちの一方がトランスミッターとしてもう一方がスキャッターとして配置されます．
## スキャッター（エネルギー基準物質）
放射光はスキャッターに入射しスキャッターからの散乱はスキャッターからの大きな立体角をカバーするためにスキャッターのすぐ上（または下）に配置された検出器によって検出されます．トランスミッターまたはスキャッターのいずれかが速度変換器に接続され，光のドップラー効果によって核共鳴エネルギーを制御します．
## 検出器
遅延した核共鳴散乱と即時の電子散乱を適切な時間窓を設定することで区別するために，サブナノ秒オーダーの高い時間応答と高い動的範囲が必要です．通常は遅延する核共鳴散乱が探索核種の励起状態の寿命に対応して放出されるため，アバランシェフォトダイオード（APD）検出器が使用されます​．

これらの構成要素は物質の超微細構造を分析し，核共鳴エネルギーを精密に制御するために組み合わされます．このようにして物質の詳細な性質や状態を非破壊的に調査することが可能になります．
