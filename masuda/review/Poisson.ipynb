# ポアソン分布の性質

『現代数理統計学の基礎』https://www.kyoritsu-pub.co.jp/book/b10003681.html

## 期待値
ポアソン分布の確率質量関数は以下のように与えられます：

$$ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} $$

ここで、$ X $ はポアソン分布に従う確率変数、$ k $ は非負の整数、$ \lambda $ は平均および分散を表すパラメータです。

期待値 $ E[X] $ は以下のように計算されます：

$$ E[X] = \sum_{k=0}^{\infty} k \cdot P(X = k) $$
$$ = \sum_{k=0}^{\infty} k \cdot \frac{e^{-\lambda} \lambda^k}{k!} $$

この式を展開し、$ k! = k \cdot (k-1)! $ という事実を利用して、

$$ E[X] = \sum_{k=1}^{\infty} \frac{e^{-\lambda} \lambda^k}{(k-1)!} $$
$$ = \lambda e^{-\lambda} \sum_{k=1}^{\infty} \frac{\lambda^{k-1}}{(k-1)!} $$

ここで、$ \sum_{k=1}^{\infty} \frac{\lambda^{k-1}}{(k-1)!} $ は $ e^{\lambda} $ のテイラー級数展開に他ならない。

したがって、

$$ E[X] = \lambda e^{-\lambda} e^{\lambda} $$
$$ = \lambda $$

従って、ポアソン分布の期待値はパラメータ $ \lambda $ に等しい。

## 分散

ポアソン分布の分散 $ \text{Var}(X) $は期待値と同様に計算できます。分散は以下のように定義されます：

$$ \text{Var}(X) = E[X^2] - (E[X])^2 $$

まず、$ E[X^2] $を計算します：

$$ E[X^2] = \sum_{k=0}^{\infty} k^2 \cdot P(X = k) $$
$$ = \sum_{k=0}^{\infty} k^2 \cdot \frac{e^{-\lambda} \lambda^k}{k!} $$

ここで、$ k^2 = k \cdot (k - 1) + k $と分解できることを利用します：

$$ E[X^2] = \sum_{k=1}^{\infty} (k \cdot (k - 1) + k) \cdot \frac{e^{-\lambda} \lambda^k}{k!} $$
$$ = \sum_{k=1}^{\infty} k \cdot (k - 1) \cdot \frac{e^{-\lambda} \lambda^k}{k!} + \sum_{k=1}^{\infty} k \cdot \frac{e^{-\lambda} \lambda^k}{k!} $$
$$ = \sum_{k=2}^{\infty} \frac{e^{-\lambda} \lambda^k}{(k-2)!} + \lambda $$

ここで、最初の項は $ \lambda^2 $を因数として外に出すことができます：

$$ E[X^2] = \lambda^2 \sum_{k=2}^{\infty} \frac{e^{-\lambda} \lambda^{k-2}}{(k-2)!} + \lambda $$
$$ = \lambda^2 e^{-\lambda} \sum_{k=2}^{\infty} \frac{\lambda^{k-2}}{(k-2)!} + \lambda $$
$$ = \lambda^2 e^{-\lambda} e^{\lambda} + \lambda $$
$$ = \lambda^2 + \lambda $$

既に知っている $ E[X] = \lambda $を使って分散を求めます：

$$ \text{Var}(X) = E[X^2] - (E[X])^2 $$
$$ = (\lambda^2 + \lambda) - \lambda^2 $$
$$ = \lambda $$

したがって、ポアソン分布の分散もパラメータ $ \lambda $に等しい。
